```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

#Read the binary file as a list of 64 bit floats (double) precision,
#then reshape the array to have 3001 rows and 2401 columns
a = np.fromfile('height1as', dtype=np.float64)
a = a.reshape(3001,2401)

#Plot the resulting array
plt.imshow(a,cmap='jet')
plt.show()
```
